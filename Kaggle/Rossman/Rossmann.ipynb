```python
train['Date']=pd.to_datetime(train['Date'])
test['Date']=pd.to_datetime(test['Date'])
```

```python
#State Holiday Manipulation

def state_holiday_bin(x):
    if x=='a' or x=='b' or x=='c':
        return 1
    else:
        return 0
train['StateHolidayBinary']=train['StateHoliday'].apply(lambda x: state_holiday_bin(x))
test['StateHolidayBinary']=test['StateHoliday'].apply(lambda x: state_holiday_bin(x))
train.drop(['StateHoliday'], axis=1,inplace=True)
test.drop(['StateHoliday'], axis=1,inplace=True)
```

```python
#filling out the NA of the store competion open since
avg_store_competition_open_since=round(np.mean(store['CompetitionOpenSinceYear']),0)
store['CompetitionOpenSinceYear'] = store['CompetitionOpenSinceYear'].fillna(np.float64(avg_store_competition_open_since))
store['Deviation_CompetitionOpenSinceYear']=store['CompetitionOpenSinceYear'] - avg_store_competition_open_since
store.drop(['CompetitionOpenSinceYear', 'CompetitionOpenSinceMonth'], axis=1,inplace=True)
```

```python
#filling out Competition Distance
store['CompetitionDistance'] = store['CompetitionDistance'].fillna(np.float64(100000.0))
```

```python
min_Promo2SinceYear=round(np.min(store['Promo2SinceYear']),0)
store['Promo2SinceYear'] = store['Promo2SinceYear'].fillna(np.float64(min_Promo2SinceYear))
store['Promo2SinceYear_edit'] = store['Promo2SinceYear'] - min_Promo2SinceYear
store.drop(['Promo2SinceYear', 'Promo2SinceWeek'], axis=1,inplace=True)
```

```python
#manipulate StoreType
store_dummies= pd.get_dummies(store['StoreType'])
store_dummies.rename(columns={'a': 'StoreType=a', 'b': 'StoreType=b', 'c': 'StoreType=c', 'd': 'StoreType=d'}, inplace=True)
store=store.join(store_dummies_train)

assortment_dummies = pd.get_dummies(store['Assortment'])
assortment_dummies.rename(columns={'a': 'Assortment=a', 'b': 'Assortment=b', 'c': 'Assortment=c'}, inplace=True)
store=store.join(assortment_dummies_train)
store.drop(['StoreType'], axis=1,inplace=True)
store.drop(['Assortment'], axis=1,inplace=True)
```

```python
#promoInterval Manipulation
def checkStr(x):
    if isinstance(x, basestring):
        return 1
    else:
        return 0

store['PromoIntervalBin']=store['PromoInterval'].apply(lambda x: checkStr(x))
store.drop(['PromoInterval'], axis=1,inplace=True)
```
